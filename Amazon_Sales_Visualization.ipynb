```python
import pandas as pd
import numpy as np


df = pd.read_csv('clean_amazon_data.csv')
df.head()
```

```python
df['rating'].mean()
```

```python
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
sns.histplot(x='rating', data= df)
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Ratings')

plt.axvline(x= df['rating'].mean(), color='r', linestyle='--', label=f'Mean Rating: {df["rating"].mean():.2f}')
plt.legend()

plt.show() #remove the metadata and just show the chart
```

```python

plt.figure(figsize=(8,6))
sns.barplot(x='category', y='rating', data=df, errorbar=None)
plt.xticks(rotation=90)
plt.show()
```

```python

sns.kdeplot(
    data=df,
    x="discounted_price",
    y="actual_price"
)
```

```python
#Bi-variate Analysis

#2 numerical variables

#Scatterplot


plt.figure(figsize=(8,6) )
sns.scatterplot(x="discounted_price", y="actual_price", hue='category' , data=df)
```

```python
#correlation
```

```python
#Average rating by Month and Year

monthly_ratings= df.groupby("month_year")['rating_filled'].mean().reset_index()
```

```python
monthly_ratings.info()
```

```python
monthly_ratings['date']= pd.to_datetime(monthly_ratings["month_year"], format='%B %Y')
```

```python
monthly_ratings= monthly_ratings.sort_values('date')
```

```python
monthly_ratings
```

```python
monthly_ratings['month_year_new'] = monthly_ratings['date'].dt.strftime('%B %Y')
```

```python
sns.lineplot(x="month_year_new", y="rating_filled",
             data=monthly_ratings, marker='o', color='navy')
plt.xticks(rotation=90)
plt.show()
```

```python

################################## DOUBTS  ####################
import matplotlib.pyplot as plt
import numpy as np

months = np.arange(1, 13)

sales_2023 = [120, 135, 150, 160, 155, 170, 180, 175, 190, 200, 210, 220]
sales_2024 = [130, 140, 155, 165, 170, 185, 195, 190, 205, 215, 225, 240]

profit_2023 = [20, 25, 30, 28, 27, 35, 40, 38, 42, 45, 48, 50]
profit_2024 = [22, 27, 32, 30, 29, 37, 43, 41, 46, 49, 52, 55]
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Subplot 1: Sales
axes[0].plot(months, sales_2023, label="Sales 2023")
axes[0].plot(months, sales_2024, label="Sales 2024")
axes[0].set_title("Monthly Sales")
axes[0].set_xlabel("Month")
axes[0].set_ylabel("Sales")
axes[0].legend()

# Subplot 2: Profit
axes[1].plot(months, profit_2023, label="Profit 2023")
axes[1].plot(months, profit_2024, label="Profit 2024")
axes[1].set_title("Monthly Profit")
axes[1].set_xlabel("Month")
axes[1].set_ylabel("Profit")
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
x = np.arange(1, 11)

y1 = x * 2
y2 = x ** 2
y3 = np.log(x)
y4 = np.sqrt(x)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Top-left
axes[0, 0].plot(x, y1)
axes[0, 0].plot(x, y2)

axes[0, 0].set_title("y = 2x")

# Top-right
axes[0, 1].plot(x, y2)
axes[0, 1].set_title("y = x²")

# Bottom-left
axes[1, 0].plot(x, y3)
axes[1, 0].set_title("y = log(x)")

# Bottom-right
axes[1, 1].plot(x, y4)
axes[1, 1].set_title("y = √x")

plt.tight_layout()
plt.show()
```
