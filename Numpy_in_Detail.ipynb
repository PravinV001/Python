```python
#

def discount_cost(rate, cost):
    """Apply a 10% discount if the rating is 4.0 or higher."""
    if rate >= 4.0:
        return cost * 0.9
    else:
        return cost
```

```python
rating = numeric_data[:, 0]
cost = numeric_data[:, 1]
```

```python
vectorized_discount_cost = np.vectorize(discount_cost)
```

```python
vectorized_discount_cost(rating, cost)
```
