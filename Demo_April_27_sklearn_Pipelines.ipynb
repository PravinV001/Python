```python
from sklearn.base import BaseEstimator, TransformerMixin
import pandas as pd
import numpy as np

# Capping --> Winsorization
# --------------------------------------------------
# Custom Outlier Capping Transformer using IQR
# --------------------------------------------------
class OutlierCapper(BaseEstimator, TransformerMixin):
    def __init__(self, factor=1.5):
        self.factor = factor

    def fit(self, X, y=None):
        X = pd.DataFrame(X)

        self.lower_bounds_ = {}
        self.upper_bounds_ = {}

        for col in X.columns:
            Q1 = X[col].quantile(0.25)
            Q3 = X[col].quantile(0.75)
            IQR = Q3 - Q1

            lower = Q1 - self.factor * IQR
            upper = Q3 + self.factor * IQR

            self.lower_bounds_[col] = lower
            self.upper_bounds_[col] = upper

        return self

    def transform(self, X):
        X = pd.DataFrame(X).copy()

        for col in X.columns:
            lower = self.lower_bounds_[col]
            upper = self.upper_bounds_[col]

            X[col] = np.where(X[col] < lower, lower, X[col])
            X[col] = np.where(X[col] > upper, upper, X[col])

        return X
```

```python
# Correlation Finder and Remover of Redundant Features!
class CorrelationFilterWithScoring(BaseEstimator, TransformerMixin):
    def __init__(self, threshold=0.9):
        self.threshold = threshold

    def fit(self, X, y=None):
        X = pd.DataFrame(X).copy()
        y = pd.Series(y)

        corr_matrix = X.corr().abs()  #corr b/w the features!

        # Correlation of each feature with target
        target_corr = X.apply(lambda col: abs(col.corr(y)))  #corr of each feature with the output

        columns_to_drop = set()

        for i in range(len(corr_matrix.columns)):
            for j in range(i):
                col1 = corr_matrix.columns[i]
                col2 = corr_matrix.columns[j]

                if corr_matrix.iloc[i, j] > self.threshold:
                    score1 = target_corr[col1]
                    score2 = target_corr[col2]

                    # Drop the weaker feature
                    if score1 < score2:
                        columns_to_drop.add(col1)
                    else:
                        columns_to_drop.add(col2)

        self.columns_to_drop_ = list(columns_to_drop)

        return self

    def transform(self, X):
        X = pd.DataFrame(X).copy()
        return X.drop(columns=self.columns_to_drop_, errors='ignore')
```

```python
num_cols = ['loan_amnt', 'term', 'int_rate', 'annual_inc', 'dti', 'delinq_2yrs',
            'inq_last_6mths', 'open_acc', 'pub_rec', 'revol_bal', 'revol_util', 'total_acc', 'sub_grade', 'emp_length']

df[num_cols].dtypes

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

steps_num_features = [('si', SimpleImputer()), ('oc', OutlierCapper()), ('ss', StandardScaler()),
 ('fs', CorrelationFilterWithScoring())]

num_pipeline = Pipeline(steps = steps_num_features)


nom_features = ['home_ownership', 'verification_status', 'purpose']
from sklearn.preprocessing import OneHotEncoder
steps_nom_features = [('si2', SimpleImputer(strategy = 'most_frequent')), ('OHE', OneHotEncoder(handle_unknown = 'ignore', sparse_output = False)), ('ss', StandardScaler()),
 ('fs', CorrelationFilterWithScoring())]
nom_pipeline = Pipeline(steps = steps_nom_features)


from sklearn.compose import ColumnTransformer
preprocessor = ColumnTransformer([('num_pipe', num_pipeline, num_cols), ('nom_pipe', nom_pipeline, nom_features)], remainder = 'drop')

from sklearn.neighbors import KNeighborsClassifier
final_pipeline = Pipeline([('preprocess', preprocessor), ('model', KNeighborsClassifier())])

X = df.drop('loan_status', axis = 1) # Fetaures
y = df['loan_status'] # Target


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)


final_pipeline.fit(X_train, y_train)



# Feature Selection
# |Corr(X_i, X_j)| > 0.9
# |Corr(X_i, Y)| --> a1
# |Corr(X_j, Y)| --> b1
# if a1 >  b1 --> Drop X_j
```

```python
def custom_bus_metric(y_true, y_pred):
  fn = ((y_true == 1) & (y_pred == 0)).sum()
  fp = ((y_true == 0) & (y_pred == 1)).sum()
  return (100*fn + 10*fp)

from sklearn.metrics import make_scorer
custom_scorer = make_scorer(custom_bus_metric, greater_is_better = False)

rcv = RandomizedSearchCV(final_pipeline, hyper_params, cv = skf, scoring = custom_scorer, n_iter = 3)
rcv.fit(X_train, y_train)
```

```python
from sklearn.preprocessing import FunctionTransformer, StandardScaler
from sklearn.compose import ColumnTransformer

df = pd.DataFrame({
    "rate": ["45%", "60%", "75%"],
    "value": [10, 20, 30]
})

# Step 1: remove %
def clean_rate(x):
    return x.str.replace('%', '').astype(float)

rate_pipeline = Pipeline([
    ("remove_percent", FunctionTransformer(lambda x: clean_rate(x.squeeze()).to_frame())),
    ("scale", StandardScaler())
])

preprocessor = ColumnTransformer([
    ("rate_proc", rate_pipeline, ["rate"])
], remainder='passthrough')

result = preprocessor.fit_transform(df)
print(result)
```
